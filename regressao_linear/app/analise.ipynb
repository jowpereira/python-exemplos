```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv(r'C:\Users\Jonathan\Desktop\regressao_linear\Base\shampoo_2.csv')
```

```python
df = df.fillna(df.rolling(3, min_periods=1).mean().shift(1))
```

```python
result = adfuller(df['sales'].dropna())
```

```python
print(result[1])
```

```python
df['diff'] = df.sales.diff(1)
```

```python
plt.plot(df['diff'])
plt.plot(np.zeros(len(df)), '--')
plt.show()
```

```python
result = adfuller(df['diff'].dropna())
print(result[1])
```

```python
plot_acf(df['sales'].dropna(), lags=20)
plt.show()
```
