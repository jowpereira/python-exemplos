```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Jonathan\Desktop\regressao_linear\Base\shampoo_2.csv')
df = df.fillna(df.rolling(3, min_periods=1).mean().shift(1))
df = df.set_index('date').sort_index()
```

```python
result = adfuller(df['sales'].dropna())
print(result[1])
```

```python
df['diff'] = df.sales.diff(1)
df.head(3)
```

```python
#plt.plot(df.sales)
plt.plot(df['diff'])
plt.plot(np.zeros(len(df)), '--')
plt.show()
```

```python
result = adfuller(df['diff'].dropna())
print(result[1])
```

```python
plot_acf(df['sales'].dropna(), lags=20, alpha=.03)
plt.show()
```
